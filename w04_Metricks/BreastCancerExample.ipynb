```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer # import data

bc = load_breast_cancer() # load data
# sklearn.datasets(load_breast_cancer) is a bunch object not a dataframe
print(bc.keys()) # check keys
print(bc.feature_names) # from keys check "feature names"
print(bc.target) # from keys check target

#print(bc.data[:5])  # First 5 rows of feature data
#print(bc.target[:5])  # First 5 target labels
```

```python
# convert "bunch object" ref: https://stackoverflow.com/questions/48769682/how-do-i-convert-data-from-a-scikit-learn-bunch-object-to-a-pandas-dataframe
# ref: https://medium.com/@merilainen.vili/how-to-convert-a-scikit-learn-dataset-to-a-pandas-dataframe-6806fa42a98e
# ------------------------ NOT A NECESARY STEP - Convert batch object to dataframe ---------------
# a bunch object can be converted to dataframe as below 
# convert features from bc.data and add headers from bc.feature_names
df = pd.DataFrame(bc.data, columns=bc.feature_names)
# add target data to dataframe from bc.target
df['target'] = bc.target
#print(df.head())
df
```

<br> y for dependent target (bc.target)

```python
print(bc.data.shape)
print(bc.target.shape, np.unique(bc.target))
print(bc.feature_names)
X = bc.data
y = bc.target
```

The breast cancer dataset:

- (569, 30): 569 samples, each with 30 features.
- (569,): 569 labels (binary classification: 0 = benign, 1 = malignant).
- [0 1]: Possible label values (benign and malignant).

```python
# Import Gaussian Naive Bayes classifier, a probabilistic model used for classification 
# based on Bayes' theorem, assuming that the features follow a Gaussian (normal) distribution.
from sklearn.naive_bayes import GaussianNB 

from sklearn.model_selection import train_test_split # splitter from training and test data
from sklearn.metrics import classification_report # metrics (precision, recall, F1-score, support) 
```

```python
# split data to train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1138) # default 25% test size
```

```python
model = GaussianNB()  # initialize a Gaussian Naive Bayes classifier model.
model.fit(X_train, y_train)  # fit data
print("Model overall score")
print(model.score(X_test, y_test))  # accuracy score on the test set.
# Prints classification metrics for the test predictions.
print("\nClassification report:")
print(classification_report(y_test, model.predict(X_test)))  
```

>Accuracy Score: 0.965034965034965: The model achieved an accuracy of approximately 96.5% on the test set.

>Classification Report: 
<br>This details the performance metrics for each class (benign = 0, malignant = 1):

- Precision: The proportion of true positive predictions among all positive predictions.

- For benign (0): 0.98 (98%)
- For malignant (1): 0.96 (96%)
- Recall: The proportion of true positive predictions among all actual positive cases.

- For benign (0): 0.92 (92%)
- For malignant (1): 0.99 (99%)
- F1-Score: The harmonic mean of precision and recall, balancing both metrics.

- For benign (0): 0.95
- For malignant (1): 0.97
- Support: The number of actual occurrences of each class in the test set.

- Benign (0): 52
- Malignant (1): 91

>Overall Metrics: 
- Accuracy: Overall accuracy across both classes is 97%.
- Macro Average: Average performance metrics without considering class imbalance.
- Weighted Average: Average performance metrics considering class imbalance (weighted by support).

## Cross Validation using KFold

Let's first of all get the validation scores without standardscaler

```python
from sklearn.model_selection import KFold
```

## IMPORTANT

With KFold always be careful with whether you want it to shuffle the folds or not. I said in lectures the folds need to be shuffled

Since we have already used train_test_split which did some shuffling for us, in this case, we don't need to shuffle again. But THIS IS NOT ALWAYS TRUE. Be aware about what earlier steps you have taken

```python
n_folds=5
ks=range(21) # go from 0 to 20 (but we'll skip 0 in the end!) you can increase this if you want
train_scores = np.zeros((len(ks),n_folds)) # creating an empty array to store the score for each fold and each degree
valid_scores = np.zeros((len(ks),n_folds))
kf = KFold(n_splits=n_folds, shuffle=False)
for k in ks:
    if k == 0:
        continue ##we can't have 0 neighbors so let's skip this one
    print(k) # Just outputting something to screen to make sure things are running

    fold = 0
    for train_index, val_index in kf.split(X_train):
        X_c, X_val = X_train[train_index], X_train[val_index]
        y_c, y_val = y_train[train_index], y_train[val_index]
        est = KNeighborsClassifier(n_neighbors = k)
        est.fit(X_c, y_c) # fit on the remaining train
        train_scores[k, fold] = est.score(X_c, y_c)
        valid_scores[k, fold] = est.score(X_val, y_val)
        fold += 1 
```

```python
mean_train_scores = train_scores.mean(axis=1)
mean_valid_scores = valid_scores.mean(axis=1)
```

```python
bestk = np.argmax(mean_valid_scores)
bestk
```

```python
mean_valid_scores[bestk]
```

## Cross Validation with StandardScaler

Now let's do validation using standardscaler and finding bestk

```python
n_folds=5
ks=range(21)
train_scores = np.zeros((len(ks),n_folds))
valid_scores = np.zeros((len(ks),n_folds))
kf = KFold(n_splits=n_folds, shuffle=False)
for k in ks:
    fold = 0
    if k == 0:
        continue ##we can't have 0 neighbors so let's skip this one
    print(k)
        
    for train_index, val_index in kf.split(X_train):
        X_c, X_val = X_train[train_index], X_train[val_index]
        y_c, y_val = y_train[train_index], y_train[val_index]
        est = make_pipeline(
            StandardScaler(),
            KNeighborsClassifier(n_neighbors = k)
        )
        est.fit(X_c, y_c) # fit on the remaining train
        train_scores[k, fold] = est.score(X_c, y_c)
        valid_scores[k, fold] = est.score(X_val, y_val)
        fold += 1 
  
```

```python
mean_train_scores = train_scores.mean(axis=1)
mean_valid_scores = valid_scores.mean(axis=1)
```

```python
bestk = np.argmax(mean_valid_scores)
bestk
```

```python
mean_valid_scores[bestk]
```

### Validation score with standard scaler and k=3 is the best one so we'll go with that

Let's plot it now just to see

```python
import matplotlib.pyplot as plt
```

```python
plt.axvline(bestk, 0,0.5, color='r', label="best valid score at k=%d"%bestk, alpha=0.3)
plt.plot(ks[1:], mean_train_scores[1:], marker='o', label='train error', alpha=0.9)
plt.plot(ks[1:], mean_valid_scores[1:], marker='o', label='CV error', alpha=0.9)
plt.legend(loc='upper right')
plt.xticks(range(1,21))
plt.xlabel('k')
plt.ylabel('accuracy score')
plt.show()
```
